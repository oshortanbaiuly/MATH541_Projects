```python
def covariance(x, y):
    mean_x = x.mean()
    mean_y = y.mean()

    sub_x = [i - mean_x for i in x]
    sub_y = [i - mean_y for i in y]

    numerator = sum([sub_x[i]*sub_y[i] for i in range(len(sub_x))])
    denominator = len(x)-1

    return numerator/denominator

covariance(data_train.iloc[:,3], data_train.iloc[:,-1])
```
